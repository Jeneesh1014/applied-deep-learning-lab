**Snippet 6**: Implement the following description based on Python classes:

- A building consists of rooms
- A building can be initialized with a list of rooms
- A building has a size and a location given as two floating point coordinates
- Each room has a number, a capacity and several properties described in natural language
- It needs to be asserted that the room number is in the range of 1 to 100
- A building allows for swapping the room numbers of two rooms


```python
from typing import List

class Room:
    def __init__(self,number,capacity,properties):
        assert 1<=number<=100, "Not Allowed"
        self.number:int= number
        self.capacity:int = capacity
        self.properties:List[str] = properties
    
    def __str__(self):
        return f"Room number: {self.number}"


class Building:
    def __init__(self,rooms,size,location):
        super().__init__()
        self.rooms:List[Room] = rooms
        self.size:int=size
        self.location:str = location

    def swap_rooms(self,room1,room2):
        room1.number,room2.number = room2.number,room1.number




room1 = Room(45,10,"Great View")
room2 = Room(46,20,"Large Hall")

building = Building([room1,room2],1400,'north-east')


print(room1)
print(room2)
building.swap_rooms(room1,room2)

print(room1)
print(room2)

```
